```python
df['Time_taken'].fillna(df['Time_taken'].mean(), inplace=True) ## Missing value treatment
```

```python
q1=df['Marketing expense'].quantile(0.25)   ## Outlier treatment
q3=df['Marketing expense'].quantile(0.75)
iqr=q3-q1
lf=q1-1.5*iqr
uf=q3+1.5*iqr
outlier_condition= (df['Marketing expense']<lf) | (df['Marketing expense']>uf)
df.loc[outlier_condition,'Marketing expense']=df['Marketing expense'].median()
df.loc[outlier_condition]     
```

```python
def outlier_treatment(data,columns):
    for col in  columns:
        q1=data[col].quantile(0.25)
        q3=data[col].quantile(0.75)
        iqr=q3-q1
        lf=q1-1.5*iqr
        uf=q3+1.5*iqr
        outlier_condition= (data[col]<lf) | (data[col]>uf)
        data.loc[outlier_condition,col]=data[col].median()
        
    return data
columns=['Marketing expense','Budget','Lead_ Actor_Rating','Lead_Actress_rating',
'Director_rating','Producer_rating','Trailer_views','Twitter_hastags','Num_multiplex','Collection']  
```

```python
data= outlier_treatment(df, columns)
```
